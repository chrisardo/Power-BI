### **<font color="#FFFFFF"> ¿Qué son las Funciones DAX? </font>**
</td>
</tr>
</table>

Las funciones DAX (Data Analysis Expressions) son un conjunto de funciones, operadores y constantes utilizadas en herramientas de análisis de datos como Power BI, Excel Power Pivot y SQL Server Analysis Services (SSAS).

Estas funciones permiten realizar cálculos avanzados y manipulaciones de datos dentro de modelos tabulares para generar informes dinámicos y paneles interactivos.

### **<font color="#FFFFFF"> Función DAX SUM </font>**
</td>
</tr>
</table>

La función SUM en DAX (Data Analysis Expressions) es una de las funciones más básicas y utilizadas para realizar cálculos en modelos de datos tabulares, como los que se encuentran en Power BI, Excel Power Pivot o SQL Server Analysis Services.

Esta función permite sumar los valores de una columna numérica específica.



```
Cantidades = SUM(FactResellerSales[OrderQuantity])
```
